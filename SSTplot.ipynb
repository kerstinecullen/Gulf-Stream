```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import xarray as xr
%matplotlib inline
```

```python
#Download latest SST data
ds=xr.open_dataset("http://psl.noaa.gov/thredds/dodsC/Datasets/noaa.oisst.v2.highres/sst.day.mean.2025.nc", engine="netcdf4")

```

```python

ds.sst
```

```python

#subset dataset
sst1 = ds.sst.isel(time=0,lat=slice(460,570),lon=slice(1100,1280)) 

#plot 
plt.clf()
fig = plt.figure(figsize = (10, 5))

axis = fig.add_subplot(1, 1, 1, projection = ccrs.PlateCarree())
sst1.plot(ax = axis,cmap='RdBu_r',vmin=2,vmax=30)

axis.coastlines('10m',)

plt.show()

```
